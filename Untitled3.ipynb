```python
!pip install gensim matplotlib scikit-learn
#import gensim
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
from gensim.models import Word2Vec
medical_corpus=[
  "The patient was diagnosed with diabetes and hypertenstion."
  "MRI scans revel abnormality in the brain tissu"
  "the treatment involves antibiotic and regular monitoring. "
  "Symptoms include fever fatigue and muscles pain"
  "The vaccine is effective against several viral infections."
  "Doctors recommend physical therapy for recovery"
  "The clinical trial result were published in the journal"
  "The surgeon performed a minimally invasive prodedure."
  "The prescription includes pain relivers and anti-inflammatory drugs."
  "THe diagnosis confirmed a rare genetic disorder."
]
processed_corpus=[sentence.lower().split() for sentence in medical_corpus]
model=Word2Vec(sentences=processed_corpus,vector_size=100,window=5,min_count=1,workers=4,epochs=50)
word=list(model.wv.index_to_key)
embeddings=np.array([model.wv[word] for word in  word])
tsne=TSNE(n_components=2,random_state=42,perplexity=5)
tsne_result=tsne.fit_transform(embeddings)
plt.figure(figsize=(10,8))
plt.scatter(tsne_result[:,0],tsne_result[:,1],color="red")
for i, word in enumerate(word):
  plt.text(tsne_result[i,0]+0.02,tsne_result[i,1]+0.02,word,fontsize=12)
plt.title("word Embeddings Visulizaton (Medical Domain)")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.grid(True)
plt.show()
def find_similar_words(input_word,topn=5):
  try:
    similar_words=model.wv.most_similar(input_word,topn=topn)
    print(f"words similar to '{input_word}':")
    for word,similarity in similar_words:
      print(f"{word}:{similarity:.2f}")
  except KeyError:
    print(f"'{input_word}' not found in the vocabulary.")
find_similar_words("treatment")
find_similar_words("vaccine")
```
